```python
bad_fip = []
def fip_to_high_school(input_num):
    result = education_data[education_data['County Fip'] == input_num]['% high school only'].tolist()
    if (len(result) == 0): 
        bad_fip.append(input_num)
        return 0.3
    else:
        return result[0]
def fip_to_college(input_num):
    result = education_data[education_data['County Fip'] == input_num]['% college or higher'].tolist()
    if (len(result) == 0): 
        bad_fip.append(input_num)
        return 0.3
    else:
        return result[0]
```

```python
pop_data['% high school only'] = np.vectorize(lambda x : fip_to_high_school(x))(pop_data['County Fip'])
```

```python
pop_data['% college or higher'] = np.vectorize(lambda x : fip_to_college(x))(pop_data['County Fip'])
```

```python
def clean_total(total_gdp):
    if (total_gdp == '(NA)'):
        return float(100)
    elif (total_gdp == '(D)'):
        return float(1)
    else:
        return float(total_gdp)
def percentage_gdp(input_num, total_gdp):
    if (input_num == '(NA)'):
        return float('nan')
    elif (input_num == '(D)'):
        return float(0)
    else:
        return (float(input_num) / total_gdp)

gdp_fip_data['total_industry_gdp'] = np.vectorize(lambda x : clean_total(x))(gdp_data.iloc[0:107990:34]['2012'])
gdp_fip_data['% total_private_gdp'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[1:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 11'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[2:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 21'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[3:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 22'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[4:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 23'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[5:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 31-33'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[6:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 321,327-339'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[7:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 311-316,322-326'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[8:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 42'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[9:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 44-45'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[10:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 48-49'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[11:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 51'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[12:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 52,53'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[13:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 52'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[14:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 53'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[15:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 54,55,56'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[16:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 54'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[17:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 55'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[18:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 56'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[19:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 61,62'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[20:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 61'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[21:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 62'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[22:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 71,72'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[23:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 71'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[24:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 72'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[25:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
gdp_fip_data['% 81'] = np.vectorize(lambda x,y : percentage_gdp(x,float(y)))(gdp_data.iloc[26:107990:34]['2012'], gdp_fip_data['total_industry_gdp'])
```
