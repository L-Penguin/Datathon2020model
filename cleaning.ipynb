## Data cleaning...

```python
import numpy as np
import pandas as pd
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("data\Part 4b.csv") 
```

```python
data = data.drop("RCPSZFE.id", 1)
```

```python
data = data.drop("YEAR.id", 1)
```

```python
data = data.drop("GEO.id", 1)
```

```python
data = data.drop(0)
```

```python
data
```

```python
data['City Name'] = np.vectorize(lambda x: x[11:-5])(data["GEO.display-label"][:])
data['State'] = np.vectorize(lambda x: x[-3:-1])(data["GEO.display-label"][:])
```

```python
data = data.drop("GEO.display-label", 1)
```

```python
data = data.rename(columns={"GEO.id2" : "zip"})
```

```python
data["zip"] = data["zip"].astype("str")
```

```python
data.to_csv("part_4a.csv") 
```
