```python
data = pd.read_csv("data2.csv", index_col = 0)
```

```python
append0 = np.vectorize(lambda x : str(x) if len(str(x)) == 5 else "0" + str(x))
data["zip"] = append0(data["zip"])
```

```python
# a = data.loc[:, ["zip", "city scale", "City Name"]]
```

```python
temp = data[data["scale"] == "Total"]
temp = temp[temp["level"] == 1]
```

```python
a  = temp.loc[:, ["zip", "city scale", "City Name"]]
```

```python
dic = {"tiny" : 1, "small" : 10, "medium" : 100, "large" : 1000}
replace = np.vectorize(lambda x : dic[x])
a["city scale"] = replace(a["city scale"])
```

```python
a = a.drop("zip", 1)
```
