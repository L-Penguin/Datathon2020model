```python
zip_info = pd.read_csv('our_data/uszips.csv')
```

```python
def zip_to_fip(zip_num):
    return zip_info[zip_info['zip'] == zip_num]['county_fip'].tolist()[0]
def zip_to_county(zip_num):
    result = zip_info[zip_info['zip'] == zip_num]['county_name'].tolist()
    if (len(result)):
        return result[0]
    else:
        return "NA"
def zip_to_county_fip(zip_num):
    result = zip_info[zip_info['zip'] == zip_num]['county_fips'].tolist()
    if (len(result)):
        return result[0]
    else:
        return float('nan')
def zip_to_lat(zip_num):
    result = zip_info[zip_info['zip'] == zip_num]['lat'].tolist()
    if (len(result)):
        return result[0]
    else:
        return float('nan')
def zip_to_lng(zip_num):
    result = zip_info[zip_info['zip'] == zip_num]['lng'].tolist()
    if (len(result)):
        return result[0]
    else:
        return float('nan')
```

```python
counties = np.vectorize(lambda x : zip_to_county(x))(org_data['zip'].unique())
```

```python
zip_curr = [0]
zip_cnt = [-1]
def assign_county(zip_num):
    if (zip_curr[0] != zip_num):
        zip_cnt[0] += 1
        zip_curr[0] = zip_num
    return counties[zip_cnt[0]]
```

```python
org_data['County Names'] = [assign_county(i) for i in org_data['zip']]
```

```python
county_fips = np.vectorize(lambda x : zip_to_county_fip(x))(org_data['zip'].unique())
```

```python
zip_curr = [0]
zip_cnt = [-1]
def assign_county_fip(zip_num):
    if (zip_curr[0] != zip_num):
        zip_cnt[0] += 1
        zip_curr[0] = zip_num
    return county_fips[zip_cnt[0]].astype('int')
```

```python
org_data['County Fips'] = [assign_county_fip(i) for i in org_data['zip']]
```

```python
county_lat = np.vectorize(lambda x : zip_to_lat(x))(org_data['zip'].unique())
zip_curr = [0]
zip_cnt = [-1]
def assign_city_lat(zip_num):
    if (zip_curr[0] != zip_num):
        zip_cnt[0] += 1
        zip_curr[0] = zip_num
    return county_lat[zip_cnt[0]].astype('float')
org_data['City Lat'] = [assign_city_lat(i) for i in org_data['zip']]
```

```python
county_lng = np.vectorize(lambda x : zip_to_lng(x))(org_data['zip'].unique())
zip_curr = [0]
zip_cnt = [-1]
def assign_city_lng(zip_num):
    if (zip_curr[0] != zip_num):
        zip_cnt[0] += 1
        zip_curr[0] = zip_num
    return county_lng[zip_cnt[0]].astype('float')
org_data['City Lng'] = [assign_city_lng(i) for i in org_data['zip']]
```
